```python
# ============================================================
# BLOQUE 1 - CLASE PERSONAJE CON CONSTRUCTOR
# ============================================================
print("========== BLOQUE 1: CLASE PERSONAJE ==========")

class Personaje:

    def __init__(self, nombre: str, clase: str,
                 vida_max: float, ataque: int, defensa: int):
        self.nombre      = nombre
        self.clase       = clase
        self.nivel       = 1           # siempre empieza en 1
        self.vida        = vida_max    # empieza con vida llena
        self.vida_maxima = vida_max
        self.ataque      = ataque
        self.defensa     = defensa

    def __str__(self):
        return (f'[{self.clase}] {self.nombre}'
                f' Nv{self.nivel}'
                f' | HP:{self.vida:.0f}/{self.vida_maxima:.0f}')

    # Ejercicio 1A: barra de vida
    def mostrar_estado(self):
        pct     = (self.vida / self.vida_maxima) * 100
        relleno = int(pct / 10)
        barra   = '#' * relleno + '.' * (10 - relleno)

        if   pct <= 0:  estado = 'MUERTO'
        elif pct <= 25: estado = 'CRITICO'
        elif pct <= 50: estado = 'HERIDO'
        elif pct <= 75: estado = 'ESTABLE'
        else:           estado = 'SALUDABLE'

        print(f'[{barra}] {pct:.0f}% - {estado}')

# ── Pruebas ──
guerrero = Personaje('Aragorn', 'Guerrero', 100, 18, 12)
mago     = Personaje('Gandalf', 'Mago',      70, 28,  5)
arquero  = Personaje('Legolas', 'Arquero',   85, 22,  8)

print("===== PERSONAJES =====")
print(guerrero)
print(mago)
print(arquero)

# Modificar uno NO afecta a los demas
guerrero.vida -= 30
print("\nDespues de recibir dano:")
print(f'Vida guerrero: {guerrero.vida}')
print(f'Vida mago:     {mago.vida}')

print("\nEstado del guerrero:")
guerrero.mostrar_estado()

print(f'\nSon objetos distintos: {guerrero is mago}')
```

```python
# ============================================================
# BLOQUE 2 - ENCAPSULAMIENTO
# ============================================================
print("========== BLOQUE 2: ENCAPSULAMIENTO ==========")

class Personaje:

    CLASES_VALIDAS = ['Guerrero', 'Mago', 'Arquero']

    def __init__(self, nombre: str, clase: str,
                 vida_max: float, ataque: int, defensa: int):
        self.__nombre      = nombre
        self.__nivel       = 1
        self.__vida_maxima = vida_max
        self.__ataque      = ataque
        self.__defensa     = defensa
        self.__vida        = vida_max
        self.clase         = clase     # usa el setter

    # ── GETTERS ──
    @property
    def nombre(self):     return self.__nombre
    @property
    def nivel(self):      return self.__nivel
    @property
    def vida(self):       return self.__vida
    @property
    def ataque(self):     return self.__ataque
    @property
    def defensa(self):    return self.__defensa

    # ── SETTERS con validacion ──

    # Vida: no puede ser negativa ni pasar del maximo
    @vida.setter
    def vida(self, v):
        if   v < 0:                   self.__vida = 0
        elif v > self.__vida_maxima:  self.__vida = self.__vida_maxima
        else:                         self.__vida = v

    # Clase: solo acepta Guerrero, Mago o Arquero
    @property
    def clase(self): return self.__clase
    @clase.setter
    def clase(self, c):
        if c in self.CLASES_VALIDAS:
            self.__clase = c
        else:
            print(f'Clase invalida: {c}. Se asigna Guerrero.')
            self.__clase = 'Guerrero'

    # Ejercicio 2A: nivel entre 1 y 50
    @nivel.setter
    def nivel(self, n):
        if   n < 1:  self.__nivel = 1
        elif n > 50: self.__nivel = 50
        else:        self.__nivel = n

    # ── METODOS DE NEGOCIO ──

    def recibir_dano(self, dano: int):
        dano_real  = max(1, dano - self.__defensa)
        self.vida -= dano_real
        print(f'{self.__nombre} recibe {dano_real} de dano. HP:{self.__vida:.0f}')

    def curar(self, cantidad: float):
        self.vida += cantidad
        print(f'{self.__nombre} curado. HP:{self.__vida:.0f}/{self.__vida_maxima:.0f}')

    def esta_vivo(self) -> bool:
        return self.__vida > 0

    def mostrar_estado(self):
        pct     = (self.__vida / self.__vida_maxima) * 100
        relleno = int(pct / 10)
        barra   = '#' * relleno + '.' * (10 - relleno)

        if   pct <= 0:  estado = 'MUERTO'
        elif pct <= 25: estado = 'CRITICO'
        elif pct <= 50: estado = 'HERIDO'
        elif pct <= 75: estado = 'ESTABLE'
        else:           estado = 'SALUDABLE'

        print(f'[{barra}] {pct:.0f}% - {estado}')

    def __str__(self):
        return (f'[{self.__clase}] {self.__nombre}'
                f' Nv{self.__nivel}'
                f' | HP:{self.__vida:.0f}/{self.__vida_maxima:.0f}')

# ── Pruebas ──
guerrero = Personaje('Aragorn', 'Guerrero', 100, 18, 12)
mago     = Personaje('Gandalf', 'Mago',      70, 28,  5)

print("===== PERSONAJES =====")
print(guerrero)
print(mago)

print("\n===== COMBATE =====")
guerrero.recibir_dano(25)
guerrero.mostrar_estado()

print("\n===== CURACION =====")
guerrero.curar(10)
guerrero.mostrar_estado()

print(f'\nGuerrero vivo: {guerrero.esta_vivo()}')

print("\n===== SETTER NIVEL =====")
guerrero.nivel = 0
print(f'Nivel con 0:  {guerrero.nivel}')   # debe ser 1
guerrero.nivel = 75
print(f'Nivel con 75: {guerrero.nivel}')   # debe ser 50
guerrero.nivel = 10
print(f'Nivel con 10: {guerrero.nivel}')   # debe ser 10

print("\n===== CLASE INVALIDA =====")
guerrero.clase = 'Dios'
print(f'Clase: {guerrero.clase}')          # debe ser Guerrero
```
